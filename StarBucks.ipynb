```python
import pandas as pd
df1=pd.read_csv(r"/content/Products_Table.csv")
df1
df2=pd.read_csv(r"/content/Orders_Table.csv")
df2
```

```python
combined_df = pd.concat([df1, df2])
combined_df
```

```python
combined_df.head(100)
combined_df.tail(100)
```

```python
# 1. Merge both datasets to create a master sales table.

Master_sales=pd.merge(df1,df2,on="product_id",how="outer")
Master_sales
```

```python

# 2. Calculate total revenue for each product.

Master_sales['Revenue']=Master_sales['unit_price'] * Master_sales['transaction_qty']
Master_sales

total_revenue = Master_sales.groupby('product_type')['Revenue'].sum().reset_index()
total_revenue=total_revenue.sort_values(by='Revenue',ascending=False)
total_revenue
```

```python
# 4. Find the top 10 most sold products.


total_revenue = Master_sales.groupby('product_type')['Revenue'].sum().reset_index()
total_revenue=total_revenue.sort_values(by='Revenue',ascending=False).head(10)
total_revenue
```

```python
# 5. Count the number of transactions for each payment mode.

Payment_Mode=Master_sales['payment_mode'].value_counts()
Payment_Mode
```

```python
# 5. Count the number of transactions for each payment mode.

import matplotlib.pyplot as plt

# Grouping and counting transactions by payment mode
payment_counts = Master_sales['payment_mode'].value_counts().reset_index()
payment_counts.columns = ['payment_mode', 'transaction_qty']

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(payment_counts['payment_mode'], payment_counts['transaction_qty'], color="orange")
plt.xlabel('Payment Mode')
plt.ylabel('Transaction Count')
plt.title("Transactions per Payment Mode")
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
